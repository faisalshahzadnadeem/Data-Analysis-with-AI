# Data Analysis with Chatgpt

```python
import pandas as pd

# Load the data
df = pd.read_csv('patients.csv')

# Display the first few rows of the dataframe
print("First few rows of the dataframe:")
print(df.head())
```

```python
# Display basic information about the dataframe
print("\nBasic information about the dataframe:")
print(df.info())
```

```python
# Fill missing values in 'Diagnosis' with 'Unknown'
df['Diagnosis'].fillna('Unknown', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count of Diagnoses
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Diagnosis', order=df['Diagnosis'].value_counts().index)
plt.title('Count of Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

# Boxplot of Lab Results by Diagnosis
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Diagnosis', y='LabResult')
plt.title('Boxplot of Lab Results by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Lab Result')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Age vs. Lab Result colored by Diagnosis
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='LabResult', hue='Diagnosis', palette='Set1')
plt.title('Scatter plot of Age vs. Lab Result colored by Diagnosis')
plt.xlabel('Age')
plt.ylabel('Lab Result')
plt.legend(title='Diagnosis', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
